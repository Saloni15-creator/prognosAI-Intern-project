```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
column_names = [
    "engine_id", "cycle",
    "op_setting_1", "op_setting_2", "op_setting_3"
] + [f"sensor_{i}" for i in range(1,22)]

df = pd.read_csv(
    '/content/cmapss_cleaned_train_FD001.csv',
    sep=',',
    header=None,
    names=column_names
)

print("Shape of the DataFrame:", df.shape)
df.head()
```

```python
df.info()
df.describe().transpose()
```

```python
print("Missing values per column:")
print(df.isnull().sum())
```

```python
if df.isnull().any().any():
    df = df.fillna(method='ffill').fillna(method='bfill')
    print("Missing values after filling:", df.isnull().sum().sum())
else:
    print("No missing values detected.")
```

```python
df.to_csv('cmapss_cleaned_train_FD001.csv', index=False)
```

```python
print(f"The cleaned dataset has {df.shape} rows and {df.shape} columns.")
print("Data loading and initial cleaning complete.")
```
